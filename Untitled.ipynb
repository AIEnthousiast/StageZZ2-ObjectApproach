```python
from SNOPSolutionOutOfPlace import SNOPSOlutionOutOfPlace
from Instances import GridInstance
import gurobipy as gp
from SNOPCompactModel import SNOPCompactModel
```

```python
instance = GridInstance(n=5)
```

```python
sol = SNOPSOlutionOutOfPlace(instance.get_random_strong_orientation())
```

```python
sol.show()
```

```python
P = list(map(lambda x : x[1], sorted(instance.get_minimum_paths().values(),key=lambda x : x[0],reverse=True)))
```

```python
env = gp.Env(empty=True)
env.setParam('OutputFlag', 0)
env.start()
model = SNOPCompactModel(env=env)
model.construct_model(instance)
```

```python
model.solve()
```

```python
N = 100
best_value = float("inf")
best_sol = None

for i in range(N):
    model.fix_path(P[i])
    meta = model.solve()
    if meta.objective_value < best_value:
        best_value = meta.objective_value
        best_sol = meta.misc["solution"]
```
